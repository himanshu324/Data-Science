```python
import numpy as np
import math
import random
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
data= pd.read_csv('data.csv')
language = list(data['LanguagesWorkedWith'])
csv_reader = Counter()
for lang in language:
    csv_reader.update(lang.split(';'))
colors = list(map(lambda i: "#" + "%06x" % random.randint(0, 0xFFFFFF),range(len(language))))
languages = []
popularity = []
for a,b in csv_reader.items():
    languages.append(a)
    popularity.append(b)
y=np.array(popularity)  
#below operation was difficult to do so we changed the popularity list into numpy
porcent = 100*y/y.sum()
  
patches,text = plt.pie(popularity, colors=colors, startangle=90, radius=2.0)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(languages, porcent)]

sort_legend = True
if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, popularity),
                                          key=lambda x: language[2],
                                          reverse=True))

plt.legend(patches, labels,loc='best',bbox_to_anchor=(1.4, 1.2),
           fontsize=8)

plt.savefig('piechart.png', bbox_inches='tight')

plt.show()


```
