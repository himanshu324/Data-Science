```python
import csv 
from collections import Counter
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
with open('data.csv') as data:
    csv_reader = csv.DictReader(data)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

print(language_counter)
language = []
popularity = []
for a,b in language_counter.items():
    
    language.append(a)
    popularity.append(b)
    
popularity.sort(reverse=True)

plt.figure(figsize=(10,15))    
plt.barh(language,popularity)
# Creating the legend of the bars in the pl
plt.title('Prograaming languages VS Number of users',color='b')
plt.ylabel('Programing languages')
plt.xlabel('Number of users')
plt.xticks(rotation=45)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_linewidth(0.5)
plt.gca().spines['left'].set_visible(True)
plt.show()
plt.tight_layout()
```
